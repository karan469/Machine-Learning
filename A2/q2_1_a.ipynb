```python
dest_folder = './fashion_mnist/'
train_filename = 'train.csv'
test_filename = 'test.csv'
val_filename = 'val.csv'

train_size = 22500
test_size = 5000
val_size = 2501
image_size = 28
label = 784
```

```python
import numpy as np
from cvxopt import matrix
```

```python
def loadCSV(filename, size):
    n = np.zeros(shape = (size, image_size*image_size + 1))
    f = open(dest_folder+filename, 'r')
    i = 0
    for x in f:
        lst = x.split(',')
        n[i] = np.asarray(lst, dtype=np.float32)
        i += 1
    for i in range(size):
        for j in range(image_size*image_size):
            n[i][j] /= 255
    return n
```

```python
n1 = loadCSV(train_filename, train_size)
n2 = loadCSV(test_filename, test_size)
n3 = loadCSV(val_filename, val_size)
```
